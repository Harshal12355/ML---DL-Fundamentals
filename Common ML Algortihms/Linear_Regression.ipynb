# Implementing Linear Regression with gradient descent

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Generate or load dataset (for simplicity, let's create some synthetic data)
# We'll create a simple linear relationship: y = 2x + 1 with some added noise
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)  # 4 is theta_0, 3 is theta_1
```

```python
theta_0 = 0.5 
theta_1 = 0.5
learning_rate = 0.001
iterations = 1000
m = len(x)
```

```python
def h(x, theta_0, theta_1):
    return theta_0 + theta_1 * x
```

```python
def compute_cost(x, y, theta_0, theta_1):
    predictions = h(x, theta_0, theta_1)
    return np.sum((predictions - y) **2) * (1/ (2 * m))
```

```python
def compute_gradients(x, theta_0, theta_1):
    predictions = h(x, theta_0, theta_1)
    d_theta_0 = (1/m) * np.sum(predictions - y)
    d_theta_1 = (1/m) * np.sum((predictions - y) * x)
    return d_theta_0, d_theta_1
```

```python
def gradient_descent(x, y, theta_0, theta_1, learning_rate, iterations):
    cost_history = []
    for iteration in range(iterations):
        d_theta_0, d_theta_1 = compute_gradients(x, theta_0, theta_1)
        theta_0 = theta_0 - (learning_rate * d_theta_0)
        theta_1 = theta_1 - (learning_rate * d_theta_1)
        cost = compute_cost(x,y, theta_0, theta_1)
        cost_history.append(cost)
        if iteration % 100 == 0:
            print(f'Iteration {iteration} the cost is : {cost:.4f}')
        
    return theta_0, theta_1, cost_history
```

```python
theta_0, theta_1, cost_history = gradient_descent(x, y, theta_0, theta_1, learning_rate, iterations)
```

```python
# Plotting the results 

# Original data points
plt.scatter(x, y, label='Data Points')
```

```python
# Plot the regression line
plt.plot(x, h(x, theta_0, theta_1), color="red", label="Regression Line")
plt.scatter(x, y, label='Original Data Points')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Fit")
plt.show()
```

```python
plt.plot(range(iterations), cost_history, color="purple")
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Cost Convergence")
plt.show()
```
